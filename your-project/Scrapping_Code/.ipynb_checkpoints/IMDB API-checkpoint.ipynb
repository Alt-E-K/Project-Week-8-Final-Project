```python
#loop to get the movie genres
for i in range(len(movies_gross.movie_id)):
    movies_info=moviesDB.get_movie(movies_gross.movie_id[i])
    try:
        genre = movies_info['genres']
        for n in range(len(genre)):
            gen=genre[n]
            gen_mov=f'{gen}-{movies_gross.movie_id[i]}'
            genres.append(gen_mov)
            print(f'{i}/{len(movies_gross.movie_id)}')
                  
    except KeyError:
        gen_mov=f'Not Av-{movies_gross.movie_id[i]}'
        genres.append(gen_mov)
        print(f'{i}/{len(movies_gross.movie_id)}')

```

```python
#we save the list for later
with open("genres.txt", "wb") as fp:   #Pickling
    pickle.dump(genres, fp)
```

```python
#loop to get the rates
for i in range(len(movies_gross.movie_id)):
    movies_info=moviesDB.get_movie(movies_gross.movie_id[i])
    try:        
        rate = movies_info['rating']
        rates.append(rate)
        print(f'{i}/{len(movies_gross.movie_id)}')
    
    except KeyError:
        rate = f'Not Av-{movies_gross.movie_id[i]}'
        rates.append(rate)
        print(f'{i}/{len(movies_gross.movie_id)}')
    

    
```

```python
#we save the list for later
with open("rates.txt", "wb") as fp:   #Pickling
    pickle.dump(rates, fp)
```

```python
#loop to get what country is the movie from
for i in range(len(movies_gross.movie_id)):
    movies_info=moviesDB.get_movie(movies_gross.movie_id[i])
    try:
        country = movies_info['countries']
        for m in range(len(country)):
            count=country[m]
            count_mov=f'{count}-{movies_gross.movie_id[i]}'
            country.append(count_mov)
        print(f'{i}/{len(movies_gross.movie_id)}')
        
    except KeyError:
        count_mov=f'Not Av-{movies_gross.movie_id[i]}'
        country.append(count_mov)
        print(f'{i}/{len(movies_gross.movie_id)}')
```

```python
#we save the list for later
with open("countries.txt", "wb") as fp:   #Pickling
    pickle.dump(country, fp)
```
