```python
#this one is valid only when all of them have the 3 percentatges > 0 so we are going to use the other more complete version
#which allows for movies to have 0% in one or 2 of the topics
orders = []
def Topic_order(dataframe):
    for i in range(976):
        topic_0=dataframe['Topic_0'][i]
        topic_1=dataframe['Topic_1'][i]
        topic_2=dataframe['Topic_2'][i]
        if topic_0 > topic_1 > topic_2:
            order= '0,1,2'
            orders.append(order)
        elif topic_0 > topic_2 >topic_1:
            order='0,2,1'
            orders.append(order)
        elif topic_1 > topic_2 > topic_0:
            order='1,2,0'
            orders.append(order)
        elif topic_1 > topic_0 > topic_1:
            order='1,0,2'
            orders.append(order)
        elif topic_2 > topic_1 > topic_0:
            order ='2,1,0'
            orders.append(order)
        else:
            order='2,0,1'
            orders.append(order)

    return orders
```

```python
#this function creates a code or tag  with the order of the topics sorted by the percentatge of the movie they represent.
# this is the main application of machine learning to the recommender, the model has discover underlying topics 
#that are not exactly interpretable to humans (unsupervised) and returned the topic composition of each movie
#we are tagging with the same code the movies that have the same prioritization of topics under the assumption 
#topics are central to the 'mood' of the film and if a user enjoyed a movie with that particular topic combination he/she
# will also enjoy others that are similar in this respect
orders = []
def Topic_order_only(dataframe):
    for i in range(976):
        topic_0=dataframe['Topic_0'][i]
        topic_1=dataframe['Topic_1'][i]
        topic_2=dataframe['Topic_2'][i]
        if topic_0 > topic_1 > topic_2:
            order= '0,1,2'
            orders.append(order)
        elif topic_0 > topic_2 > topic_1:
            order='0,2,1'
            orders.append(order)
        elif topic_1 > topic_2 > topic_0:
            order='1,2,0'
            orders.append(order)
        elif topic_1 > topic_0 > topic_2:
            order='1,0,2'
            orders.append(order)
        elif topic_2 > topic_1 > topic_0:
            order ='2,1,0'
            orders.append(order)
        elif topic_2 > topic_0 > topic_1:
            order='2,0,1'
            orders.append(order)
        elif topic_0 == topic_1 == 0:
            order='2'
            orders.append(order)
        elif topic_1==topic_2 == 0:
            order='0'
            orders.append(order)
        else:
            order='1'
            orders.append(order)
        

    return orders
```

```python
#This is the basic model of the recomender function which only takes into account the topic codes matching, 
#it returns a list of all the movies that match.
def Recommender(df):
    movie = input('Choose a movie that you enjoyed:')
    print(f'{movie}')
    movie_row = df[df['title']== movie]
    movie_index = movie_row.index
    index_num=movie_index[0]
    reference_order = df['Order'].iloc[index_num]
    rec_list= list(df['title'][df['Order'] == reference_order])
    rec_list.remove(movie)
    return rec_list
```

```python
#Second version of the recomender
def Improved_Recommender_rating(df):
   
    #we input the movie name
    movie = input('Choose a movie that you enjoyed:')
    
    #it finds the row for that movie and grabs the index
    movie_row = df[df['title']== movie]
    movie_index = movie_row.index
    index = movie_index[0]
    #it gets the topic code for the movie and searches for all the movies that match it ( as before)
    reference_order = df['Order'].iloc[index]
    rec_frame= df[['title','rating','genres']][df['Order']==reference_order]
    rec_frame.drop(movie_index, axis=0, inplace=True)
    
    #we get to choose if we want it to work like the previous version or add the  rate filter
    active_rating_filter=input('Do you care about the movie rating?')
    
    #if we add it we can choose a number and it wil select only the movies with a rating equal or higher to the selected one 
    #and return a sorted dataframe
    if active_rating_filter == 'Yes':
    
        rating=float(input('Set your rating grade:'))
    
        rec_frame_rate = rec_frame[['title','rating','genres']][rec_frame['rating'] >= rating].sort_values('rating', ascending=False)
        return rec_frame_rate
    
    else:
        return rec_frame
```

```python
def Improved_Recommender_Rating_Genre(df):
    
    #user inpts movie
    movie = input('Choose a movie that you enjoyed:')
    
    
    
    #function gets the row in the DF then the index object, and the index number
    movie_row = df[df['title']== movie]
    movie_index_obj = movie_row.index
    movie_index=movie_index_obj[0]
    
    #funtion gets the order code for the movie with the index and generates a DF with the movies that are similar.
    reference_order = df['Order'].iloc[movie_index]
    rec_frame= df[['title','rating','genres']][df['Order']==reference_order]
    rec_frame.drop(movie_index, axis=0, inplace=True)
    rec_frame=rec_frame.reset_index()
    
    #asking weather ratings matter to filter
    active_rating_filter=input('Do you care about the movie rating?')
    
    #case if it does
    if active_rating_filter == 'Yes':
        
        #asking weather both ratings and genre matter, conditioned to ratings matter
        active_genre_filter=input('Do you care about movie genre?') 
        
        #case if genre doesnt matter provided that rates do
        if active_genre_filter=='No':
    
            rating=float(input('Set your rating grade:'))
            
            #creates a dataframe with the filter applied
            rec_frame_rate = rec_frame[['title','rating','genres']][rec_frame['rating'] >= rating].sort_values('rating', ascending=False)
            rec_frame_rate = rec_frame_rate.reset_index()
            return rec_frame_rate
        
        #case if both genre and rating matter provided rating does
        else:
            
            rating=float(input('Set your rating grade:'))
            
            #goes through the same process as previous node but a second layer is added
            rec_frame_rate = rec_frame[['title','rating','genres']][rec_frame['rating'] >= rating].sort_values('rating', ascending=False)
            rec_frame_rate = rec_frame_rate.reset_index()
            
            #asking for genres
            genre_names=['Drama','Fantasy','Action', 'Thriller','Family','Comedy','Romance', 'Horror','Animation','Western','Sci','Music','Not Av']
            print(genre_names)
            gen_input=input('Select from the previous genres: ')
            
            #iterating through every row in the column genre and transforming every cell to a list (it was read as string)
            scores=[]
            for i in range(len(rec_frame_rate['genres'])):
    
                points=0
                genres_movie=rec_frame_rate['genres'][i]
                list_gens=ast.literal_eval(genres_movie)
                
                #iterating through every item in the list, adding a point if it is among the ones the user wants
                for j in range(len(list_gens)):
                    if list_gens[j] in gen_input:
                        points +=1
                #creating a score to classify how well it matches user preferences
                score=(points/len(gen_input))*100
                scores.append(score)
        #we add that column and sort the DF with both rating and scores parameters
            rec_frame_rate['scores']=scores
            rec_frame_rate_gen=rec_frame_rate
            return rec_frame_rate_gen.sort_values(['scores', 'rating'], ascending=[False, False])
    
    else:
        #we repeat the genre process but for the DF withour rating filter
        active_genre_filter=input('Do you care about movie genre?')
                        
        if active_genre_filter=='Yes':
            genre_names=['Drama','Fantasy','Action', 'Thriller','Family','Comedy','Romance', 'Horror','Animation','Western','Sci','Music','Not Av']
            print(genre_names)
            gen_input=input('Select from the previous genres: ')
            scores=[]
            for i in range(len(rec_frame['genres'])):
    
                points=0
                genres_movie=rec_frame['genres'][i]
                list_gens=ast.literal_eval(genres_movie)
                for j in range(len(list_gens)):
                    if list_gens[j] in gen_input:
                        points +=1
                score=(points/len(gen_input))*100
                scores.append(score)
        
            rec_frame['scores']=scores
            rec_frame_gen=rec_frame
            return rec_frame_gen.sort_values('scores', ascending= False)
        else:
            return rec_frame
```

```python
def Robust_Recommender(df):
    
    movies = list(df['title'])
    #user inpts movie
    movie = input('Choose a movie that you enjoyed:')
    
    while movie not in movies:
        
        print('Choose from the following list & mind the case, please:')
        print(movies)
        movie = input('Choose a movie that you enjoyed:')
    
    else:
    
    #function gets the row in the DF then the index object, and the index number
        movie_row = df[df['title']== movie]
        movie_index_obj = movie_row.index
        movie_index=movie_index_obj[0]
    
        #funtion gets the order code for the movie with the index and generates a DF with the movies that are similar.
        reference_order = df['Order'].iloc[movie_index]
        rec_frame= df[['title','rating','genres']][df['Order']==reference_order]
        rec_frame.drop(movie_index, axis=0, inplace=True)
        rec_frame=rec_frame.reset_index()
    
        #asking weather ratings matter to filter
        active_rating_filter=input('Do you care about the movie rating? Yes or No:')
        active_rating_filter=active_rating_filter.lower()
        #case if it does
        if active_rating_filter == 'yes':
        
            #asking weather both ratings and genre matter, conditioned to ratings matter
            active_genre_filter=input('Do you care about movie genre? Yes or No:') 
            active_genre_filter=active_genre_filter.lower()
        
            #case if genre doesnt matter provided that rates do
            if active_genre_filter=='no':
    
                rating=input('Set your rating grade:')

                while rating.isdigit() == False:
                    print('Please input a number')
                    rating=input('Set your rating grade:')
                else:
            
                    rating=float(rating)
            
            #creates a dataframe with the filter applied
                    rec_frame_rate = rec_frame[['title','rating','genres']][rec_frame['rating'] >= rating].sort_values('rating', ascending=False)
                    rec_frame_rate = rec_frame_rate.reset_index()
                    return rec_frame_rate
            
            
            
            #case if both genre and rating matter provided rating does
            else:
            
                rating=input('Set your rating grade:')

                while rating.isdigit() == False:
                    print('Please input a number')
                    rating=input('Set your rating grade:')
                else:
            
                    rating=float(rating)
            
            
                    #goes through the same process as previous node but a second layer is added
                    rec_frame_rate = rec_frame[['title','rating','genres']][rec_frame['rating'] >= rating].sort_values('rating', ascending=False)
                    rec_frame_rate = rec_frame_rate.reset_index()
            
                    #asking for genres
                    genre_names=['Drama','Fantasy','Action', 'Thriller','Family','Comedy','Romance', 'Horror','Animation','Western','Sci','Music','Not Av']
                    print(genre_names)
                    gen_input=input('Select from the previous genres & mind the case: ')
            
            
                    #iterating through every row in the column genre and transforming every cell to a list (it was read as string)
                    scores=[]
                    for i in range(len(rec_frame_rate['genres'])):
    
                        points=0
                        genres_movie=rec_frame_rate['genres'][i]
                        list_gens=ast.literal_eval(genres_movie)
                
                        #iterating through every item in the list, adding a point if it is among the ones the user wants
                        for j in range(len(list_gens)):
                            if list_gens[j] in gen_input:
                                points +=1
                                #creating a score to classify how well it matches user preferences
                        score=(points/len(gen_input))*100
                        scores.append(score)
                    #we add that column and sort the DF with both rating and scores parameters
                    rec_frame_rate['scores']=scores
                    rec_frame_rate_gen=rec_frame_rate
                    return rec_frame_rate_gen.sort_values(['scores', 'rating'], ascending=[False, False])
    
        else:
                #we repeat the genre process but for the DF withour rating filter
                active_genre_filter=input('Do you care about movie genre? Yes or No:') 
                active_genre_filter=active_genre_filter.lower()
        
                 #case where the rating is not important but the genre is       
                if active_genre_filter=='yes':
                    genre_names=['Drama','Fantasy','Action', 'Thriller','Family','Comedy','Romance', 'Horror','Animation','Western','Sci','Music','Not Av']
                    print(genre_names)
                    gen_input=input('Select from the previous genres: ')
           
                    scores=[]
                    for i in range(len(rec_frame['genres'])):
    
                        points=0
                        genres_movie=rec_frame['genres'][i]
                        list_gens=ast.literal_eval(genres_movie)
                        
                        for j in range(len(list_gens)):
                            if list_gens[j] in gen_input:
                                points +=1
                        score=(points/len(gen_input))*100
                        scores.append(score)
        
                    rec_frame['scores']=scores
                    rec_frame_gen=rec_frame
                    return rec_frame_gen.sort_values('scores', ascending= False)
                
                #case where neither is important
                else:
                    return rec_frame
```
