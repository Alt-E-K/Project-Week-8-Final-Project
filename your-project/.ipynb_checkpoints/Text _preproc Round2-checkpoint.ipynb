# New Stop-words

After try 1 in EDA & modelling, we go back to preprocessing to improve our model further, by substracting more words that we consider noise.

```python
import nltk
from nltk.corpus import stopwords
stopwords_list = stopwords.words("english")
stopwords_list2=stopwords_list+terms_to_drop
```

```python
list(movies_preproc['lemma_plot2'][0])
```

```python
def stopword2(x):
    output = []
    for w in x:
        if w not in stopwords_list2:
            output.append(w)
    return output
```

```python
movies_preproc['lemma_plot2'] = movies_preproc['lemma_plot2'].apply(stopword2)
```

```python
movies_preproc
```

```python
plot_df2=pd.DataFrame({'title':movies_preproc.title,
         'lemmas':movies_preproc.lemma_plot2}, columns = ['title','lemmas'])
```

```python
plot_df2
```

```python
plot_df2['lemmas'] = plot_df2['lemmas'].apply(lambda x: (' ').join([word for word in x]))
string=plot_df2['lemmas'][0]
string
```

We are trying the new (unsaved) preprocesed lemma2 with the basic vectorizer.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv_5000= CountVectorizer(max_features=5000)
data_cv = cv_5000.fit_transform(plot_df2.lemmas)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv_5000.get_feature_names())
data_dtm.index = plot_df2.index
data_dtm
```

```python
num_mentions = data_dtm.sum()
terms = data_dtm.columns
```

```python
summarized = pd.DataFrame({'term':terms,'times':num_mentions}, columns= ['term','times'])
summarized.sort_values(by=['times'],ascending=False).head(50)
```
