```python
list_directors = []
for i in range(len(splits)-1):
    if splits[i][0]!='Not Av':
        if splits[i][1]==splits[i+1][1]!=splits[i+2][1]:
            if splits[i]!=splits[i-1]:
                dir_tup=(splits[i][0],splits[i+1][0])
                list_directors.append(dir_tup)
    #except ListIndexOutOfRange
        
        elif splits[i][1]==splits[i+1][1]==splits[i+2][1]:
            if splits[i][1]!=splits[i-1][1]:
                dir_ls=[splits[i][0],splits[i+1][0],splits[i+2][0]]
                list_directors+=dir_ls
        else:
            if splits[i][1]==splits[i-1][1]:
                next
            else:
                dir_tup=(splits[i][0])
                list_directors.append(dir_tup)
    else:
        dir_tup=(splits[i][0],splits[i][1])
        list_directors.append(dir_tup)
```

```python
len(list_directors)
```

```python
with open("clean_directors.txt", "wb") as fp:   #Pickling
    pickle.dump(list_directors, fp)
```

```python
merged.rating=merged.rating.apply(floater)
```
